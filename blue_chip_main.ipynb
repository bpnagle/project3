```python
import pandas as pd
from pathlib import Path
```

```python
# load in files
data_to_load_1 = Path("Resources/AAPL.csv")
data_to_load_2 = Path("Resources/NFLX.csv")
data_to_load_3 = Path("Resources/META.csv")
data_to_load_4 = Path("Resources/GOOG.csv")
data_to_load_5 = Path("Resources/AMZN.csv")
data_to_load_6 = Path("Resources/MSFT.csv")
    
# read in data and store in df's
aapl = pd.read_csv(data_to_load_1)
nflx = pd.read_csv(data_to_load_2)
meta = pd.read_csv(data_to_load_3)
goog = pd.read_csv(data_to_load_4)
amzn = pd.read_csv(data_to_load_5)
msft = pd.read_csv(data_to_load_6)
```

```python
msft
```

```python
# data cleaning: drop adj close columns

tickers = ['aapl', 'nflx', 'meta', 'goog', 'amzn', 'msft']

for ticker in tickers:
    globals()[ticker] = globals()[ticker].drop('Adj Close', axis=1)
```

```python
# add change column 

for ticker in tickers:
    globals()[ticker]['Change'] = globals()[ticker]['Close'] - globals()[ticker]['Open']
```

```python
# add percent change column 

for ticker in tickers:
    globals()[ticker]['Percent_Change'] = ( globals()[ticker]['Change'] / globals()[ticker]['Open'] ) *100
    
```

```python
msft
```

```python
# truncate new columns

for ticker in tickers:
    globals()[ticker]['Open'] = globals()[ticker]['Open'].round(2)
    globals()[ticker]['High'] = globals()[ticker]['High'].round(2)
    globals()[ticker]['Low'] = globals()[ticker]['Low'].round(2)
    globals()[ticker]['Close'] = globals()[ticker]['Close'].round(2)
    globals()[ticker]['Change'] = globals()[ticker]['Change'].round(2)
    globals()[ticker]['Percent_Change'] = globals()[ticker]['Percent_Change'].round(2)
```

```python
# check
msft
```

```python
# drop volume and change columns

for ticker in tickers:
    globals()[ticker] = globals()[ticker].drop('Volume', axis=1)
    globals()[ticker] = globals()[ticker].drop('Change', axis=1)
```

```python
msft
```

```python
# export cleaned data as csv files
import os

output_folder = 'output_files'

# create the output folder if doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for ticker in tickers:

    df = globals()[ticker]
    
    # export the df to a csv in output folder
    df.to_csv(os.path.join(output_folder, f'{ticker}_output.csv'), index=False)
```
