```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame
ethereum_df = pd.read_excel('Resources/Ethereum.xlsx')
ethereum_df.head(30)
```

```python
# Get a brief summary of the crowdfunding_info DataFrame.
ethereum_df.info()
```

```python
# Get the crowdfunding_info_df columns.
ethereum_df.columns
```

```python
ethereum_df['date'] = pd.to_datetime(ethereum_df['Date'])

# Specify the start and end dates
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2023-12-21')

# Filter the data based on the date range
filtered_df = ethereum_df[(ethereum_df['Date'] >= start_date) & (ethereum_df['Date'] <= end_date)]

# Drop unwanted columns
ethereum_df=filtered_df.drop(["date"],axis =1)
ethereum_df=filtered_df.drop(["Currency"],axis =1)


# Print the filtered data
print(ethereum_df)
```

```python
# Get the crowdfunding_info_df columns.
ethereum_df.columns
```

```python
# Export the selected dataframe to a new CSV file
ethereum_df.to_csv('ethereum_df.csv', index=False)
```
