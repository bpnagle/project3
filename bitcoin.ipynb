```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame
bitcoin_df = pd.read_excel('Resources/BTC-USD.xlsx')
bitcoin_df.head(30)
```

```python
# Get a brief summary of the crowdfunding_info DataFrame.
bitcoin_df.info()
```

```python
# Get the crowdfunding_info_df columns.
bitcoin_df.columns
```

```python
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])

# Specify the start and end dates
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2023-12-21')

# Filter the data based on the date range
filtered_df = bitcoin_df[(bitcoin_df['Date'] >= start_date) & (bitcoin_df['Date'] <= end_date)]

# Drop unwanted columns
bitcoin_df=filtered_df.drop(["Adj Close", "Volume"],axis =1)


# Print the filtered data
print(bitcoin_df)
```

```python
# Export the selected dataframe to a new CSV file
bitcoin_df.to_csv('bitcoin_df.csv', index=False)
```
