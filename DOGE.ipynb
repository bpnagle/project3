```python
import petl as etl
```

```python
# Read the data into a Pandas DataFrame
#csv_file_path = 'DOGE-USD.csv'
table = etl.fromcsv('Resources/DOGE.csv')
etl.look(table)
```

```python
#Define Date Range
start_date = '2019-01-01'
end_date = '2022=08-23'
filtered_rows = etl.select (table, lambda row: start_date <= row['Date'] <= end_date)
etl.look (filtered_rows)
```

```python
filtered_rows = etl.convert(filtered_rows, {'Open':  float, 'Close': float})

filtered_rows = etl.convert(filtered_rows, {'Open':  float, 'Close': float})
filtered_rows = etl.addfield(filtered_rows, 'Percent_Change', lambda row: (row['Close'] - row['Open']) / row['Close'] *100)
#filtered_rows = etl.cutout(table, 'Adj Close', 'Volume')
etl.look (filtered_rows)
```

```python
filtered_rows = etl.selectnotnone(filtered_rows, 'Percent_Change')
etl.look(filtered_rows)
```

```python
etl.tocsv(filtered_rows, 'DOGE.csv')
```
